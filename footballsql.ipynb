```python
import pandas as pd
from pandasql import sqldf
df_players = pd.read_csv("C:/Data Science/football/players_clean.csv")
df_stats = pd.read_csv("C:/Data Science/football/stats_clean.csv")
run = lambda q: sqldf(q, globals())
```

```python
df_info = df_players.merge(df_stats, on="player_id", how='left')
```

```python
run("SELECT * FROM df_info LIMIT 5")
```

```python
run("SELECT COUNT(*) AS total_rows FROM df_info")
```

```python
run("SELECT COUNT(DISTINCT nationality) AS unique_nationality FROM df_info")
```

```python
run("SELECT player_name, SUM(goals) AS Goals FROM df_info GROUP BY player_name ORDER BY goals DESC LIMIT 10")
```

```python
run("SELECT current_club, SUM(Goals) AS club_goals FROM df_info GROUP BY current_club ORDER BY club_goals DESC LIMIT 10")
```

```python
run("SELECT player_name, SUM(goals) AS Goals, SUM(assists) AS Assists FROM df_info GROUP BY player_name HAVING assists > goals ORDER BY assists DESC")
```

```python
run("SELECT player_name, SUM(goals) AS Goals, SUM(assists) AS Assists FROM df_info GROUP BY player_name HAVING SUM(goals) > 5 AND SUM(assists) > 5 ORDER BY assists DESC")
```

```python
run("""
SELECT position,
    ROUND(AVG(goals), 2)          AS average_goals,
    ROUND(AVG(assists), 2)        AS average_assists,
    ROUND(AVG(minutes_played), 2) AS average_minutes_played
FROM df_info
GROUP BY position
ORDER BY position
""")
```

```python
run("""
SELECT position,
    AVG(goals), 2          AS average_goals,
    AVG(assists), 2        AS average_assists,
    AVG(minutes_played), 2 AS average_minutes_played
FROM df_info GROUP BY position ORDER BY position
""")
```

```python
run("""
SELECT nationality,
    ROUND(AVG(height_cm), 2) AS avg_height
FROM df_info GROUP BY nationality ORDER BY avg_height DESC
""")
```

```python
run("""
SELECT current_club,
    COUNT(player_name) AS player_count
FROM df_info
GROUP BY current_club
ORDER BY player_count DESC
""")
```

```python
run("""
SELECT preferred_foot,
    SUM(goals)   AS total_goals,
    SUM(assists) AS total_assists
FROM df_info GROUP BY preferred_foot ORDER BY preferred_foot
""")
```
