```python
def sum_of_squares(n):
    total = 0
    for i in range(1, n):
        total += i * i
    return total

sum_of_squares(4)
```

```python
def square_sum(n):
    return sum(i*i for i in range(1, n))

square_sum(4)    
```

```python
def sum_of_odd_squares(n):
    total = 0
    for i in range(1, n, 2):
        total += i * i
    return total
sum_of_odd_squares(5)
```

```python
def square_odd(n):
    return sum(i*i for i in range(1, n, 2))

square_odd(5)
```
