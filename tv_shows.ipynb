```python
# splitting each string with comma
def split_string(x):
    return x.split(",")
```

```python
#  display all actors on new row
actors = df[df["cast"].notnull()]["cast"].apply(split_string).explode()
```

```python
actors
```

```python
actors.value_counts()
```

```python
# filter out null values in cast
df_filtered = df[df["cast"].notnull()]
```

```python
# display all titles in which the actor has participated
def titles_actor(actor):

    def is_actor_in_cast(cast):
        return actor in cast
    
    return ", ".join(df_filtered[df_filtered["cast"].apply(is_actor_in_cast)]["title"].tolist())
```

```python
df2 = pd.DataFrame(actors.unique()).head(30)
df2.columns = ["actors"]
df2["title"] = df2["actors"].apply(titles_actor) 
```
